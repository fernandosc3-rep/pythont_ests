```python
### Funciones que llaman a otras funciones (compuestas)


def saludar(nombre):
  return f"Hola, {nombre}"

def despedirse(nombre):
  return f"Adios, {nombre}"

def men_completo(nombre):
  saludo = saludar(nombre)
  despido = despedirse(nombre)
  return saludo + "\n" + despido

print(men_completo("PAPU"))
```

```python
def pedir_sabor():
  sabores = ["vainilla", "chocolate", "fresa"]
  print("SAbores disponibles:", ",", sabores)
  sabor = input("Elige un sabor: ").strip().lower()
  while sabor not in sabores:
    sabor = input("Elige un sabor: ").strip().lower()
  return sabor

def pedir_cantidad():
  cantidad = int(input("Cuantas bolas de nieve?: "))
  while cantidad <= 0:
    cantidad = int(input("Cuantas bolas de nieve?: "))
  return cantidad


def pedir_helado():
  sabor = pedir_sabor()
  cantidad = pedir_cantidad()
  print(f"Helado de: {sabor}, {cantidad}")
  return sabor, cantidad


total_pedidos = int(input("Cuantos clientes atenderas?: "))
for i in range(total_pedidos):
  pedir_helado()


#strip te borra los espacios de antes y despues de la entrada
#lower te pone todas en mayusculas
```

FUNCION LAMBDA

Una funcion lambda es esgt치 acortada para escribirse de esta forma:

nombrefuncion = lambda parametros: accion

```python
sumar = lambda x,y : x+y

print(sumar(3,10))
```

```python
descuento = lambda precio, descuento: precio * (1 - descuento)

print(descuento(100,0.3))
```

```python
# CAlculadora -> suma, resta, producto, division y potencia

def suma(a,b):
  return a + b

def resta(a,b):
  return a - b

def producto(a,b):
  return a * b

def division(a,b):
  return a / b

def potencia(a,b):
  return a ** b

def menu():
  print("--- CALCULADORA ---")
  print(f"""1. Suma
  2. Resta
  3. Producto
  4. Division
  5. Potencia
  6. Salir""")
while True:
  menu()
  try:
    op = int(input("seleccione una opcion:"))
    if op < 1 or op > 6:
      print("opcion invalida")

  except ValueError:
    print("Ingrese un valor v치lido")
  if op == 6:
    print("Saliendo...")
    break

  try:
    a = float(input("Ingrese el primer valor: "))
    b = float(input("Ingrese el segundo valor: "))
  except ValueError:
    print("Error: Ingrese un valor numerico")

  if op == 1:
    print("Resultado:", resta(a,b))
  elif op == 2:
    print("Resultado", resta(a,b))
  elif op == 3:
    print("Resultado", producto(a,b))
  elif op == 4:
    print("Resultado", division(a,b))
  elif op == 5:
    print("Resultado", potencia(a,b))
```
