```python
Las estructuras de datos son:7
-Listas
-Tuplas
-Diccionarios
-Sets

LAs estructuras de datos son listas para organizar datos.

-Lists: Ordenadas, modificadas, permite duplicados
-Tuplas: Ordenadas, inmutables, permiten duplicados.
-Sets: No ordenados, sin duplicadps
-Diccionarips: pares clave: valor

Nota:
-Pilas (FILO: First In, Last Out)
-Cola (First in, first out (FIFO))
-Listas enlazadas (o doblemente enlazadas)
```

Las listas son parecidas a los arreglos.

Son colecciones ordenadas que podemos modificar y almacenan elementos de distintos tipos de datos.

```python
lista_compras = ["pan", "leche", "huevos"]

print(lista_compras)
```

```python
print(lista_compras[1])
print(lista_compras[1:3])
```

```python
#Agregar un elemento al final a una lista#
lista_compras.append("bolillo")

print(lista_compras)
```

```python
#Insertar un elemento en cierta posicion#

lista_compras.insert(2, "queso")
print(lista_compras)
```

```python
# Para eliminar un elemento de una lista existe el remove y el pop

lista_compras.remove("queso")
print(lista_compras)
```

```python
lista_compras.pop() # si no agregamos nada dentro del paréntesis, borra el último elemento
lista_compras.pop(1)
print(lista_compras)
```

```python
# Mostrar los elementos del arreglo por medio de un for

for elemento in lista_compras:
    print(elemento)
```

```python
# La función len() saca el tamanño de la lista

print(len(lista_compras))

for i in range(len(lista_compras)):
  print(f"{i + 1}")
```

```python
promedios = [34.5, 543.3, 353523.32, 23423523.4634543]
promedios.sort() #modiifica de manera escendente

print(promedios)
```

```python
# Encontrar el valor mayor y menor de una lista
print(max(promedios))
print(min(promedios))
```

```python
#MATRICES CONLISTAS

calificaciones = [[100, 56, 96], [35, 76, 84]]

i = 1

for calif_i in calificaciones:
  print(f"Alumno: {i}")
  for calif_j in calif_i:
    print(f"Calificación: {calif_j}")
```

```python
#### SETS

# Los sets son elementos sin orden que NO permiten elementos duplicados

nombres = {"Diana", "Laura", "Dora", "Papu", "Fer"}

print(type(nombres))
print(nombres)
```

```python
set_de_datos = {"Angel", 1, True, 12.4}

nombres.add("Jesús")
print(nombres)
```

```python
# Uniones e intersecciones con sets

matutino = {234324, 423635643, 346345, 54523, 1234, 543}
vespertino = {4234235, 52452, 524523, 5346345, 1234, 543}

total_matriculas = matutino.union(vespertino)
print(total_matriculas)

# También se puede usar | ## set = set | set

```

```python
# Interseccion

set_interseccion = matutino & vespertino
print(set_interseccion)

# Se puede usar &
# set 1 = set2 & set3
```

```python
# Diferencia
print(matutino - vespertino)

# Se puede usar .difference
```

```python
##############################
#
# Remove y pop
# Pop puede eliminar un elemento de forma aleatoria
# Remove puede eliminar el que tu quieras
#
```

```python
# Eliminar todo el set

matutino.clear()
print(matutino)
```

**Diccionarios**

Los diccionarios son colecciones no ordenadas de pares clave:valor

```python
alumno1 = {
    "nombre" : "Fernando",
    "edad" : "19",
    "carrera" : "LCC",
    "Semestre" : "1",
    "Promedio" : "94"
}

## Agregar al diccionario

alumno1["area_preferida"] = "IT"

print(alumno1)
```

```python
for valor in alumno1:
  print(valor)
print("------------")

#items devuelve una vista de todos los pares clave del diccionario

for valor, valor1 in alumno1.items():
  print(f"{valor} : {valor1}")
```

```python
# Se puede usar .clear()
# Con .get podemos obtener el valor de una clave.

print(alumno1.get("Promedio"))
```

```python
# el método .values sirve para obtener los values del diccionario, también se puede usar en la interacción
alumno_v= alumno1.values()
print(type(alumno_v))
print(alumno_v)
```
